```python
a = 2
b = 2.0
c = a

print(f"a: {a}, type: {type(a)}")
print(f"b: {b}, type: {type(b)}")
print(f"c: {c}, type: {type(c)}")
print("---")

print(f"a == b: {a == b}")
print(f"a is b: {a is b}")
print("Explanation: 'a' and 'b' have the same numerical value but are different types (int vs. float) and therefore different objects in memory.")
print("---")

print(f"a == c: {a == c}")
print(f"a is c: {a is c}")
print("Explanation: 'c' was assigned 'a', so they refer to the exact same integer object.")
print("---")

print(f"b == c: {b == c}")
print(f"b is c: {b is c}")
print("Explanation: 'b' and 'c' have the same numerical value, but 'b' is a float and 'c' is an integer, making them different objects.")
```

```python
total = 0
count = 0

while True:
    num_str = input('Enter a number: ')
    if num_str == 'done':
        break

    try:
        num = float(num_str)
    except ValueError:
        print('Invalid input')
        continue

    total += num
    count += 1

if count > 0:
    average = total / count
    print(f'{int(total)} {count} {average}')
else:
    print('No numbers were entered.')
```

```python
largest = None
smallest = None

while True:
    num_str = input('Enter a number: ')
    if num_str == 'done':
        break

    try:
        num = float(num_str)
    except ValueError:
        print('Invalid input')
        continue

    if largest is None or num > largest:
        largest = num
    if smallest is None or num < smallest:
        smallest = num

if largest is not None and smallest is not None:
    print('Maximum:', largest)
    print('Minimum:', smallest)
else:
    print('No numbers were entered.')
```
