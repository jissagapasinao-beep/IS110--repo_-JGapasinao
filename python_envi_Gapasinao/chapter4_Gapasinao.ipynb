```python
def computepay(hours, rate):
    if hours > 40:
        # Calculate regular pay for the first 40 hours
        regular_pay = 40 * rate
        # Calculate overtime hours
        overtime_hours = hours - 40
        # Calculate overtime pay
        overtime_pay = overtime_hours * (rate * 1.5)
        # Total pay is regular pay plus overtime pay
        total_pay = regular_pay + overtime_pay
    else:
        # If no overtime, total pay is simply hours multiplied by rate
        total_pay = hours * rate
    return total_pay

hours_str = input('Enter Hours: ')
rate_str = input('Enter Rate: ')

try:
    hours = float(hours_str)
    rate = float(rate_str)
    pay = computepay(hours, rate)
    print('Pay:', pay)
except:
    print('Error: Please enter numeric input for hours and rate')
```

```python
def computegrade(score):
    if not (0 <= score <= 100):
        return 'Bad score'
    elif 97 <= score <= 100:
        return '1.0'
    elif 93 <= score < 97:
        return '1.25'
    elif 89 <= score < 93:
        return '1.5'
    elif 85 <= score < 89:
        return '1.75'
    elif 81 <= score < 85:
        return '2.0'
    elif 77 <= score < 81:
        return '2.25'
    elif 74 <= score < 77:
        return '2.5'
    elif 0 <= score < 74:
        return '3.0'
    else:
        return 'Bad score'

score_str = input('Enter score (between 0 and 100): ')

try:
    score = float(score_str)
    grade = computegrade(score)
    print('Grade:', grade)
except:
    print('Bad score')
```
